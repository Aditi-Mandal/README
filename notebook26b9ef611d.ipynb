```python
#Import Required Libraries

import pandas as pd
import numpy as np
import matlotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

#Load the dataset

df = pd.read_csv("Mall_Customers.csv")
df.head()

#Data processing

df.isnull().sum()
df.fillna(df.mean(),inplace=True)
df['Gender'] = df['Gender'].map({'Male':1,'Female':0})
x = df[['Annual Income(k$)','Spending Score(1-100)']]
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

#Determine the optimal number of cluster(K)

wcss = []
for k in range(1,11):
    kmeans = kmeans(n_clusters=k,init='k-means++',random_state=42)
    kmeans.fit(x_scaled)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(range(1,11),wcss,marker='0',linestyle='--')
plt.xlabel('Number of clusters(k)')
plt.ylabel('wcss')
plt.title('Elbow Method')
plt.show()

#Apply k-Means Clustering

kmeans = kmeans(n_clusters=5,init='k-means++',random_state=42)
clusters = kmeans.fit_predict(x_scaled)
df['cluster'] = clusters

#Visualize Clusters

plt.figure(figsize=(8,6))
sns.scatterplot(x=df['Annual Income(k$)'],y=df['spending score(1-100)'],hue=df['Cluster'],palette='virdis')
plt.scatter(kmeans.cluster_centers_[:,0]*scaler.scale_[0]+scaler.mean_[0],
            kmeans.cluster_centers_[:,1]*scaler.scale_[1]+scaler.mean_[1,]
            s=300,c='red',label='Centroids',marker='X')
plt.xlabel("Annual Income(k$)")
plt.ylabel("Spending Score (1-100)")
plt.title("Customer Segmentation using k-Means")
plt.legend()
plt.show()

#Interpret the segments
1.High-income,high-spending customers=>Prenium customers.
2.High-income,low-spending customers=>Price-sensitive customers.
3.Low-income,high-spending customers=>Young spenders.
4.Low-income,low-spending customers=>Budget-conscious customer.
5.Mid-income,moderate-spending customers=>Average buyers.
```
